```python
import pandas as pd
import json

```

```python
# Đọc dữ liệu từ file Excel
df = pd.read_excel('F:/Vlancer_Project/241.Luan_van_Chatbot_using_LSTM/data/Form_ud_20240404_23h33.xlsx', sheet_name='summary')
```

```python
df
```

```python
# Tạo một danh sách các dictionary để lưu trữ thông tin về mỗi nhóm câu hỏi
data = []

# Duyệt qua từng dòng trong dataframe
current_tag = None
patterns = []
responses = []
for index, row in df.iterrows():
    tag = row['Nhóm câu hỏi']
    question = row['Câu hỏi']
    answer = row['Câu trả lời']
    
    # Nếu tag mới, lưu các patterns và responses của tag trước vào data
    if current_tag is not None and tag != current_tag:
        data.append({'tag': current_tag, 'patterns': patterns, 'responses': responses})
        patterns = []
        responses = []
    
    # Lưu patterns và responses vào danh sách tương ứng
    patterns.append(question)
    responses.append(answer)
    current_tag = tag

# Lưu nhóm câu hỏi cuối cùng vào data
data.append({'tag': current_tag, 'patterns': patterns, 'responses': responses})

for item in data:
    item['patterns'] = [x for x in item['patterns'] if not pd.isnull(x)]
    item['responses'] = [x for x in item['responses'] if not pd.isnull(x)]
```

```python
data_vf = {"intents": data}
```
